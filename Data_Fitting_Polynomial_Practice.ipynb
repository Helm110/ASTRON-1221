```python
#Import required Libraries
import numpy as np # Import numpy to use in calculations
import astropy.constants as const #Package contains the astonomy constants used later
import astropy.units as u # import a Python package to call certain units, e.g., mass of the sun
import math #Import math to use in calculations
import matplotlib.pyplot as plt #Import to plot graphs
```

```python
#Define information array
Income_array = np.array([36700,46400,50200,54200,75100,89200])
Education_Leveln = np.arange(len(Income_array))
Education_Levels = np.array(["Less Than Highschool","Highschool", "Some College","Associates Degree","Bachelors Degree","Post Graduate and Beyond"])

#Print out arrays for bug fixing
#print(Income_array)
#print(Education_Level)

#Plot Array
plt.plot(Education_Levels, Income_array, 'ro')
plt.xlabel("Education Level (Degree Level)")
plt.ylabel("Income ($)")
plt.xticks(rotation=45)
plt.title("Average Income vs Education Level for Men")
plt.show()
```

```python
#Use polyfit to find a line of best fit to approx. data

#Linear Approx.
linFit = np.polyfit(Education_Leveln, Income_array, 1)
#Test to understand what the value is
print(linFit)
#Polynomial Approx.
quadFit = np.polyfit(Education_Leveln, Income_array, 2)
#Test to understand what the value is
print(quadFit)
```

```python
#Define Error Bars
error = np.array([3110,1070,2000,4130,4110,8910])

#Defining Equations
linear = linFit[0]*Education_Leveln + linFit[1]
quadratic = quadFit[0]*Education_Leveln**2 + quadFit[1]*Education_Leveln + quadFit[2]

#Plot linear & polynomial approx with data points
plt.errorbar(Education_Leveln, Income_array, yerr=error, fmt='ro',color = "Black")
plt.plot(Education_Leveln, linear, color = "blue", label = "Linear Fit")
plt.plot(Education_Leveln, Income_array, color = "red", label = "Quadratic Fit")
plt.xlabel("Education Level (Degree Level)")
plt.ylabel("Income ($)")
plt.title("Average Income vs Education Level for Men")
plt.show()
```

```python
#Calculate Chi to compare the two models
n = len(Income_array);
m_linear = 2;
m_quad = 3;
chi_linear = np.sum(((Income_array - linear)**2/error**2)/ (n - m_linear));
chi_quad = np.sum(((Income_array - quadratic)**2/error**2)/ (n - m_quad));
#Print out the results of each
print("The Reduced Chi2 foe the linear fit is = {:.2f}".format(chi_linear))
print("The Reduced Chi2 foe the quadratic fit is = {:.2f}".format(chi_quad))
```

```python
print("This result shows that the linear fit is a better representation of our data")
```
