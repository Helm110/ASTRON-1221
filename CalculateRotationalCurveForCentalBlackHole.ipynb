```python
# prompt: can you calculate the change in orbital velocity as the radius increases from the black hole that is central to the milkyway
"""
This code is used as a reference and a base to understanding the flow of code
It is also used to check simple errors like spelling mistakes, and incorrect use of np, const, and u
"""
import numpy as np
import astropy.constants as const #Package contains the astonomy constants used later
import astropy.units as u # import a Python package to call certain units, e.g., mass of the sun

# Constants
G = 6.67430e-11  # Gravitational constant (m^3 kg^-1 s^-2)
M = 4.1e6 * 1.989e30  # Mass of Sagittarius A* (kg)

def orbital_velocity(radius):
  """Calculates the orbital velocity at a given radius from the black hole.

  Args:
    radius: Distance from the black hole in meters.

  Returns:
    Orbital velocity in meters per second.
  """
  return np.sqrt(G * M / radius)

# Example usage:
radius1 = 1e12  # Initial radius (1 trillion meters)
radius2 = 2e12  # Increased radius (2 trillion meters)

velocity1 = orbital_velocity(radius1)
velocity2 = orbital_velocity(radius2)

change_in_velocity = velocity2 - velocity1

print("Orbital velocity at radius 1:", velocity1, "m/s")
print("Orbital velocity at radius 2:", velocity2, "m/s")
print("Change in orbital velocity:", change_in_velocity, "m/s")
print(G)
```

```python
def calcOrbitV(M,r):
    """
    This is the function to calculate orbital velocity
    Input:
    M, the mass of the central object
    r, the distance between the two objects
    Output:
    orbitialVelocity, the orbital velocity
    """
    G = const.G.to(u.km**3 / u.s**2 / u.kg)
    orbitalVelocity =  np.sqrt(G*M/r)
    return orbitalVelocity

```

```python
#Defining the inputs M & r in respects to Earth
M = 1*u.M_earth
r = 1*u.earthRad

#Runs computation of OrbitalVelocity function
v_earthSatalite = calcOrbitV(M,r)
print(v_earthSatalite.to(u.km/u.s))
```

```python
#Defining an array and it's applications
# Mercury, Venus, Earth, Mars, Jupiter, Saturn, Uranus, Neptune

orbital_distance_au = [0.39, 0.72, 1, 1.52, 5.20, 9.58, 19.2, 30.1]* u.au
```

```python
# Define orbital Velocities for all planets
M_sun = 1*u.M_sun
v_planet_arr = calcOrbitV(M_sun,orbital_distance_au)
print(v_planet_arr.to(u.km/u.s))
```
